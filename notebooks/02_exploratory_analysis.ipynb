## Exploratory Data Analysis

##### The Variables
- Wind Speed (metres per second)
- Pitch Angle (degrees of rotation from perpendicular to the oncoming wind)
- Power Output (kilowatts (kW) of power generated)

##### Understanding the Data:
1. What is the distribution of wind speeds?
2. What is the distribution of power output?
3. How does wind relate to power?
4. What does pitch angle do?
5. How does generation change over time?

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# loading the data
df = pd.read_csv('../data/processed/turbine_clean.csv')
```

### 1. What is the distribution of wind speeds?

```python
# generating a histogram of wind speeds
sns.histplot(df['wind_speed'], kde=True)
plt.show()
df['wind_speed'].describe()
```

The mean wind speed over the period is $6.19 m/s$. There is a standard deviation of $2.78 m/s$ and the distribution has a rightward skew. Regular spikes of values at certain regular increments could refer to some rounding issues concentrating multiple values to those points.

### 2. What is the distribution of power output?

```python
# generating a histogram of power output
sns.histplot(df['power_output'], kde=False)
plt.show()
df['power_output'].describe()
```

The mean power output for the period is 653.85 kW. There is a consistent downward trend across the range of power ouputs, with exceptions at the tails. There is a spike around 2000 kW which is due to the Vestas V100 Turbine being rated to a 2 mW maximum capacity. There is also a spike at 0 including some negative values (data minimum is -27.5). These points would be cause for investigation.
